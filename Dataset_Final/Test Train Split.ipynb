```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("/content/gdrive/My Drive/VIT/Tamil Argumentation/Dataset/Merged/merged_data.xlsx")
```

```python
df.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size=0.25, random_state=42)
```

```python
train_data = pd.DataFrame(X_train, columns=['Sno', 'Dataset', 'Topic', 'Tweet and Comments', 'Language', 'Quality',
       'Stance wrt topic', 'Stance wrt content', 'Argument', 'Comment',
       'Responding to Tone', 'Discussing Writer Characteristics', 'Remark'])

train_data['Relevancy'] = y_train

test_data = pd.DataFrame(X_test, columns=['S No', 'Date of Tweet', 'Topic', 'Parent Tweet', 'Tweet',
       'Stance wrt topic', 'Stance wrt content', 'Language', 'Quality',
       'Argument', 'Comment', 'Responding to Tone',
       'Discussing Writer Characteristics', 'Remark'])

test_data['Relevancy'] = y_test

# Save DataFrames to CSV files
train_data.to_excel('/content/gdrive/My Drive/VIT/Tamil Argumentation/Dataset/Twitter/Train/twitter_data_train.xlsx', index=False)
test_data.to_excel('/content/gdrive/My Drive/VIT/Tamil Argumentation/Dataset/Twitter/Test/twitter_data_test.xlsx', index=False)
```
